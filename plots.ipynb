```python
import matplotlib.pyplot as plt

from sklearn import metrics
import pandas as pd


"""
read the plots that PNN.py made
make histograms that compare the two performances

"""

NN73 = pd.read_csv('m700_m300.csv', sep=r'\s')
#PNN = pd.read_csv('PNN.csv', sep=r'\t')
PNN73 = pd.read_csv('PNN73.csv', sep=r'\t')


PNN_auc = metrics.auc(PNN73['fpr'], PNN73['tpr'])
NN73_auc = metrics.auc(NN73['fpr'], NN73['tpr'])


#plt.plot(PNN['fpr'], PNN['tpr'], label = "trained and tested without m1 = 700, m2 = 300 ")
plt.plot(NN73['fpr'], NN73['tpr'], label = f" NN trained only m1 = 700, m2 = 300, auc : {NN73_auc:.2f} ")
plt.plot(PNN73['fpr'], PNN73['tpr'], label = f" PNN trained w/o m1 = 700, m2 = 300 auc : {PNN_auc:.2f}")



plt.xlabel("FPr")
plt.ylabel("TPr")
plt.title("ROC")
plt.legend()
plt.savefig("PNN_NN_roc.pdf")
plt.show()

```
